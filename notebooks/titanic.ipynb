```python
from sklearn.ensemble import RandomForestRegressor
# ä½¿ç”¨ RandomForestClassifier å¡«è¡¥ç¼ºå¤±çš„å¹´é¾„å±žæ€§


def set_missing_ages(df):
    # æŠŠå·²æœ‰çš„æ•°å€¼åž‹ç‰¹å¾å–å‡ºæ¥ä¸¢è¿›Random Forest Regressorä¸­
    age_df = df[['Age', 'Fare', 'Parch', 'SibSp', 'Pclass']]
 
    # ä¹˜å®¢åˆ†æˆå·²çŸ¥å¹´é¾„å’ŒæœªçŸ¥å¹´é¾„ä¸¤éƒ¨åˆ†
    known_age = age_df[age_df.Age.notnull()].as_matrix()
    unknown_age = age_df[age_df.Age.isnull()].as_matrix()
 
    # yå³ç›®æ ‡å¹´é¾„
    y = known_age[:, 0]
 
    # Xå³ç‰¹å¾å±žæ€§å€¼
    X = known_age[:, 1:]
 
    # fitåˆ°RandomForestRegressorä¹‹ä¸­
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
    rfr.fit(X, y)
 
    # ç”¨å¾—åˆ°çš„æ¨¡åž‹è¿›è¡ŒæœªçŸ¥å¹´é¾„ç»“æžœé¢„æµ‹
    predictedAges = rfr.predict(unknown_age[:, 1:])
 
    # ç”¨å¾—åˆ°çš„é¢„æµ‹ç»“æžœå¡«è¡¥åŽŸç¼ºå¤±æ•°æ®
    df.loc[(df.Age.isnull()), 'Age'] = predictedAges 
 
    return df, rfr
 
 
def set_Cabin_type(df):
    df.loc[(df.Cabin.notnull()), 'Cabin'] = "Yes"
    df.loc[(df.Cabin.isnull()), 'Cabin'] = "No"
    return df
 
 
data, rfr = set_missing_ages(data)
data = set_Cabin_type(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.learning_curve import learning_curve
 
 
# ç”¨sklearnçš„learning_curveå¾—åˆ°training_scoreå’Œcv_scoreï¼Œä½¿ç”¨matplotlibç”»å‡ºlearning curve
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=1, 
                        train_sizes=np.linspace(.05, 1., 20), verbose=0, plot=True):
    """
    ç”»å‡ºdataåœ¨æŸæ¨¡åž‹ä¸Šçš„learning curve.
    å‚æ•°è§£é‡Š
    ----------
    estimator : ä½ ç”¨çš„åˆ†ç±»å™¨ã€‚
    title : è¡¨æ ¼çš„æ ‡é¢˜ã€‚
    X : è¾“å…¥çš„featureï¼Œnumpyç±»åž‹
    y : è¾“å…¥çš„target vector
    ylim : tupleæ ¼å¼çš„(ymin, ymax), è®¾å®šå›¾åƒä¸­çºµåæ ‡çš„æœ€ä½Žç‚¹å’Œæœ€é«˜ç‚¹
    cv : åšcross-validationçš„æ—¶å€™ï¼Œæ•°æ®åˆ†æˆçš„ä»½æ•°ï¼Œå…¶ä¸­ä¸€ä»½ä½œä¸ºcvé›†ï¼Œå…¶ä½™n-1ä»½ä½œä¸ºtraining(é»˜è®¤ä¸º3ä»½)
    n_jobs : å¹¶è¡Œçš„çš„ä»»åŠ¡æ•°(é»˜è®¤1)
    """
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, verbose=verbose)
 
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
 
    if plot:
        plt.figure()
        plt.title(title)
        if ylim is not None:
            plt.ylim(*ylim)
        plt.xlabel(u"è®­ç»ƒæ ·æœ¬æ•°")
        plt.ylabel(u"å¾—åˆ†")
        plt.gca().invert_yaxis()
        plt.grid()
 
        plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, 
                         alpha=0.1, color="b")
        plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, 
                         alpha=0.1, color="r")
        plt.plot(train_sizes, train_scores_mean, 'o-', color="b", label=u"è®­ç»ƒé›†ä¸Šå¾—åˆ†")
        plt.plot(train_sizes, test_scores_mean, 'o-', color="r", label=u"äº¤å‰éªŒè¯é›†ä¸Šå¾—åˆ†")
 
        plt.legend(loc="best")
 
        plt.draw()
        plt.show()
        plt.gca().invert_yaxis()
 
    midpoint = ((train_scores_mean[-1] + train_scores_std[-1]) + (test_scores_mean[-1] - test_scores_std[-1])) / 2
    diff = (train_scores_mean[-1] + train_scores_std[-1]) - (test_scores_mean[-1] - test_scores_std[-1])
    return midpoint, diff
 

plot_learning_curve(clf, u"å­¦ä¹ æ›²çº¿", X, y)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import sklearn.preprocessing as preprocessing
from sklearn import linear_model, cross_validation

def set_missing_ages(df):
    """
    ä½¿ç”¨éšæœºæ£®æž—å¡«è¡¥ageç©ºç¼º
    :param df: ç©ºç¼ºæ•°æ®é›†
    :return: å¡«è¡¥åŽçš„æ•°æ®é›†ï¼Œéšæœºæ£®æž—æ¨¡åž‹
    """
    age_df = df[['Age', 'Fare', 'Parch', 'SibSp', 'Pclass']]  # ç”¨äºŽæ‹Ÿåˆå¹´é¾„çš„å­—æ®µ

    # ä¹˜å®¢åˆ†æˆå·²çŸ¥å¹´é¾„å’ŒæœªçŸ¥å¹´é¾„ä¸¤éƒ¨åˆ†
    known_age = age_df[age_df.Age.notnull()].values
    unknown_age = age_df[age_df.Age.isnull()].values

    # è¾“å‡º
    y = known_age[:, 0]

    # è¾“å…¥ç‰¹å¾
    X = known_age[:, 1:]

    rfr = RandomForestRegressor(random_state=0, n_estimators=1000, n_jobs=-1)
    '''
    å‚æ•°ï¼š
        - random_state 
            è®©ç»“æžœå®¹æ˜“å¤çŽ°ã€‚åœ¨å‚æ•°å’Œè®­ç»ƒæ•°æ®ä¸å˜çš„æƒ…å†µä¸‹ï¼Œä¸€ä¸ªç¡®å®šçš„éšæœºå€¼å°†ä¼šäº§ç”Ÿç›¸åŒçš„ç»“æžœ
        - n_estimators å­æ¨¡åž‹æ•°é‡
        - n_jobs å¤„ç†é™åˆ¶
            - -1: ä¸é™åˆ¶
            -  n: nä¸ªå¤„ç†å™¨
         
    '''
    rfr.fit(X, y)

    # ç”¨å¾—åˆ°çš„æ¨¡åž‹è¿›è¡ŒæœªçŸ¥å¹´é¾„ç»“æžœé¢„æµ‹
    predicted_ages = rfr.predict(unknown_age[:, 1::])

    # ç”¨å¾—åˆ°çš„é¢„æµ‹ç»“æžœå¡«è¡¥åŽŸç¼ºå¤±æ•°æ®
    df.loc[(df.Age.isnull()), 'Age'] = predicted_ages

    return df, rfr


def set_cabin(df):
    """
    å¤„ç†cabin
    :param df:
    :return:
    """
    df.loc[(df.Cabin.notnull()), 'Cabin'] = "Yes"
    df.loc[(df.Cabin.isnull()), 'Cabin'] = "No"
    return df


def numerical(df):
    """
    æ•°å€¼åŒ–éžæ•°å€¼å­—æ®µ
    :param df:
    :return:
    """
    dummies_Cabin = pd.get_dummies(df['Cabin'], prefix='Cabin')
    dummies_Embarked = pd.get_dummies(df['Embarked'], prefix='Embarked')
    dummies_Sex = pd.get_dummies(df['Sex'], prefix='Sex')
    dummies_Pclass = pd.get_dummies(df['Pclass'], prefix='Pclass')

    df = pd.concat([df, dummies_Cabin, dummies_Embarked, dummies_Sex, dummies_Pclass], axis=1)
    df.drop(['Pclass', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)

    return df


def scaling(df):
    """
    ç¼©æ”¾Ageå’ŒFare
    :param df:
    :return:
    """
    scaler = preprocessing.StandardScaler()

    Age = df.Age.values.reshape(-1, 1)  # fitéœ€è¦äºŒç»´å‘é‡
    Fare = df.Fare.values.reshape(-1, 1)

    age_scale_param = scaler.fit(Age)
    df['Age_scaled'] = scaler.fit_transform(Age, age_scale_param)
    fare_scale_param = scaler.fit(Fare)
    df['Fare_scaled'] = scaler.fit_transform(Fare, fare_scale_param)

    return df


def preprocessor(df):
    """
    é¢„å¤„ç†æ•°æ®é›†
    :param df:
    :return:
    """
    df, rfr = set_missing_ages(df)
    df = set_cabin(df)
    df = numerical(df)
    df = scaling(df)

    return df


def train(df):
    """
    è®­ç»ƒæ¨¡åž‹
    :param df:
    :return:
    """
    # ç”¨æ­£åˆ™å–å‡ºæˆ‘ä»¬è¦çš„å±žæ€§å€¼
    train_df = df.filter(regex='Survived|Age_.*|SibSp|Parch|Fare_.*|Cabin_.*|Embarked_.*|Sex_.*|Pclass_.*')
    train_np = train_df.values

    # yå³Survivalç»“æžœ
    y = train_np[:, 0]

    # Xå³ç‰¹å¾å±žæ€§å€¼
    X = train_np[:, 1:]

    clf = linear_model.LogisticRegression(C=1.0, penalty='l1', tol=1e-6)
    clf.fit(X, y)


train_df = pd.read_csv('data/train.csv')
test_df = pd.read_csv('data/test.csv')

# set_missing_ages(train_df)
# set_cabin(train_df)
# train_df = numerical(train_df)
train_df = preprocessor(train_df)
train_df.info()
```
